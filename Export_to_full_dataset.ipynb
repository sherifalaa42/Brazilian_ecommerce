```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer = pd.read_csv("D:/Projects2/Brazilian E-Commerce/olist_customers_dataset.csv")
geoloc = pd.read_csv("D:/Projects2/Brazilian E-Commerce/olist_geolocation_dataset.csv")
order_items = pd.read_csv("D:/Projects2/Brazilian E-Commerce/olist_order_items_dataset.csv")
order_payments = pd.read_csv("D:/Projects2/Brazilian E-Commerce/olist_order_payments_dataset.csv")
order_reviews = pd.read_csv("D:/Projects2/Brazilian E-Commerce/olist_order_reviews_dataset.csv")
orders =  pd.read_csv("D:/Projects2/Brazilian E-Commerce/olist_orders_dataset.csv")
products = pd.read_csv("D:/Projects2/Brazilian E-Commerce/olist_products_dataset.csv")
sellers = pd.read_csv("D:/Projects2/Brazilian E-Commerce/olist_sellers_dataset.csv")
category_name = pd.read_csv("D:/Projects2/Brazilian E-Commerce/product_category_name_translation.csv")
```

```python
datasets_list = {"customer": customer,
"geoloc": geoloc,
"order_items": order_items,
"order_payments": order_payments,
"order_reveiws": order_reviews,
"orders": orders,
"products": products,
"sellers": sellers,
"category_name": category_name}
```
